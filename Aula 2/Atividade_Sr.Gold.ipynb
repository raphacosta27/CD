## 1.Análise Teste

```python
#ANÁLISE TESTE
#Qual a proporção de respostas por região do Brasil?

regiao = dados.REGIAO.value_counts(True)
plot = regiao.plot(kind = 'pie', autopct='%.1f')
```

## 2.Como é o ensino no Brasil?

```python
dados.V6002 = dados.V6002.astype('category')
dados.V6002.cat.categories = ('Pública', 'Privada')

ensino1 = dados.V6002.value_counts(True)

plot1 = ensino1.plot(kind='pie', autopct='%.1f')
plt.title('Rede de ensino no Brasil')
plt.legend(loc='best')
```

```python
print('Tabela Cruzada 1: Cenário Brasileiro em relação à rede de ensino')
ct1 = pd.crosstab(dados.V6002, dados.REGIAO, margins=True, rownames=['Tipo da Rede de Ensino'], colnames=['Região do País'])
print(ct1)

ct1=ct1.iloc[0:2,0:5]

plot1 = ct1.plot(kind='bar', stacked=True, figsize=(8,6))
plt.grid()
plt.ylabel('Número de pessoas que frequentam')
plt.title('\nCenário brasileiro em relação à rede de ensino\n')
```

Pode-se notar que a área em que o Sr. Gold deseja atuar, está em menor número no país. Sabendo isso, o próximo passo é entender como a rede privada esta disposta no cenário brasileiro em relação ao tipo de estudante.


```python
dados.V4802 = dados.V4802.astype('category')
dados.V4802.cat.categories = ('Maternal', 'Alfabetização', 'Alfabetização Adultos', '1ª a 4ª séries', '5ª a 8ª séries', '1ª série em 9 anos', '2ª a 5ª séries em 9 anos', '6ª a 9ª séries em 9 anos', 'Jovens e adultos não EF', 'EM + pré-vest', 'superior')

ensino_privado = dados.V4802[dados.V6002 == 'Privada']
ct2 = pd.crosstab(ensino_privado, dados.REGIAO, margins=False, rownames=['Tipo da Rede de Ensino'], colnames=['Região do País'])

ct2 = ct2.loc['EM + pré-vest']

plot2 = ct2.plot(kind='pie', autopct="%.1f", figsize=(6,6))
plt.title('\nDistribuição da Rede Privada de Ensino por região\nde alunos no ensino médio e pré vestibular\n')
```

Considerando que no início, o Sr. Gold deveria ter alunos pagantes em sua universidade, fora analisado aqueles que já frenquentam o ensino privado, e disparadamente, os locais que mais tem alunos do ensino médio e pré-vestibular são, o Nordeste e Sudeste.

#### Neste momento, vemos que os melhores locais para se realizar o investimento seria o Nordeste ou o Sudeste.
