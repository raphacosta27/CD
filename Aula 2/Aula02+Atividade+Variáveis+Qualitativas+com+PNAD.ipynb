```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
notas = pd.read_table("pes2013A.txt", sep=" ")
```

```python
Norte = notas[(notas.UF > 10) & (notas.UF < 18)].iloc[:,0:1]
Sul = notas[(notas.UF >= 41) & (notas.UF <= 43)].iloc[:,0:1]
Nordeste = notas[(notas.UF >= 21) & (notas.UF <= 29)].iloc[:,0:1]
Sudeste = notas[(notas.UF >= 31) & (notas.UF <= 35)].iloc[:,0:1]
CO = notas[(notas.UF >= 50) & (notas.UF <= 53)].iloc[:,0:1]

soma_total = notas.UF.sum()

norte_relativa = (Norte.sum()/soma_total)*100
sul_relativa = (Sul.sum()/soma_total)*100
nordeste_relativa = (Nordeste.sum()/soma_total)*100
sudeste_relativa = (Sudeste.sum()/soma_total)*100
CO_relativa = (CO.sum()/soma_total)*100

lista_plots = [norte_relativa, sul_relativa, nordeste_relativa, sudeste_relativa, CO_relativa]
nomes = ["Norte", "Sul", "Nordeste", "Sudeste", "Centro-Oeste"]
plt.pie(lista_plots, labels = nomes, autopct = '%1.1f%%')
```
