```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
gdp = pd.read_excel("indicator gapminder gdp_per_capita_ppp.xlsx") #PIB per capta
```

```python
urban = pd.read_excel("indicator urban population.xlsx") #População urbana
```

```python
expvida = pd.read_excel("indicator life_expectancy_at_birth.xlsx") #Expectativa de vida
```

```python
expvida = expvida[pd.notnull(expvida["Life expectancy"])]
expvida = expvida.loc[29,:] #Expectativa de vida no Brasil
```

```python
gdp = gdp[pd.notnull(gdp['GDP per capita'])] # remove paises nao disponiveis
gdp_br = gdp.loc[29,:]
gdp_br = gdp_br.iloc[1:]
gdp_br = pd.to_numeric(gdp_br) #Transformando a Coluna em Números pois estavam como objetos
```

```python
#agua.loc[:,2010]
urban = urban[pd.notnull(urban["Urban population"])] # remove paises nao disponiveis
urban_br = urban.loc[29,:]
urban_br = urban_br.iloc[1:]
urban_br = pd.to_numeric(urban_br) #Transformando a Coluna em Números pois estavam como objetos
```

```python
expvida_br = expvida.iloc[161:213] #1960 até 2011, valores que possuimos de população urbana
```

```python
gdp_br = gdp_br.iloc[160:212] #1960 até 2011, valores que possuimos de população urbana
```

```python
plt.scatter(gdp_br, expvida_br)
plt.title("PIB per capita x Expectativa de vida")
plt.xlabel("PIB per capita")
plt.ylabel("Expectativa de vida")
plt.show()
plt.scatter(urban_br, expvida_br)
plt.title("Tamanho população urbana x Expectativa de vida")
plt.xlabel("Tamanho da população urbana")
plt.ylabel("Expectativa de vida")
plt.show()
```
