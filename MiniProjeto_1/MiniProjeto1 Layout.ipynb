### Itens 7 a 9 - Resolução:

# 6

```python
valores_normal = stats.norm.fit(jardas)
print("Normal e Laplace (Utilizam os mesmos parâmetros):")
print("Valor da média estimada: {0}".format(valores_normal[0]))
print("Valor do desvio padrão estimado: {0}".format(valores_normal[1]))
```

Portanto, a média "chutada" foi de 350 e a média real era de ~326. Assim como o desvio padrão chutado foi de 80, enquanto o verdadeiro é de ~89.

# 7 

```python
x = np.arange(10,600,0.01)
fdpj = stats.norm.pdf(x, loc=jardas.mean(), scale=jardas.std())
len(x)
fig = plt.figure(figsize=(12,8))
jardas.plot.hist(bins=22, normed=1, label = 'Jardas')
plt.plot(x, fdpj, linewidth=5, color="red", label='Normal')
plt.plot(x, stats.laplace.pdf(x, loc=jardas.mean(), scale=jardas.std()), color="green", linewidth=5, label='Laplace')
plt.legend(loc="best")
```

# 8

```python
freq_acumulada = []
for i in range (len(jardas)):
    fr = ((i-0.5)/len(jardas))
    if fr < 0:
        fr = 0
    freq_acumulada.append(fr)
freq_acumulada
```

```python
freq_acumulada = pd.Series(data=freq_acumulada)
```

```python
q_normal = stats.norm.ppf(freq_acumulada, loc=jardas.mean(), scale=jardas.std())
q_laplace = stats.laplace.ppf(freq_acumulada, loc=jardas.mean(), scale=jardas.std())
```

```python
fig = plt.figure(figsize=(12,8))
plt.title("Q-Q plot: Jardas x Normal x Laplace")
plt.plot((min(jardas), max(jardas)), (min(jardas), max(jardas)), '-', linewidth= 3, label='Jardas Reais')
norm_plot = plt.plot(jardas, q_normal, 'o', color = 'red', label='Valores - Normal')
laplace_plot = plt.plot(jardas, q_laplace, 'o', color = 'green', label='Valores - Laplace')
plt.legend(loc='best')
```

# 9

```python
cdf_normal = stats.norm.cdf(jardas, scale=jardas.std(), loc=jardas.mean()) #FDA
cdf_laplace = stats.laplace.cdf(jardas, scale=jardas.std(), loc=jardas.mean())

fig = plt.figure(figsize=(12,8))

plt.title("Cumulativa: Jardas x Normal x Laplace")

jardas_fra_plot, = plt.plot(jardas, freq_acumulada, '-', color='black', label="Fr dos dados")
cdf_normal_plot, = plt.plot(jardas, cdf_normal, '-', color='red', label = "Cumulativa da normal")
cdf_laplace_plot, = plt.plot(jardas, cdf_laplace, '-', color='green', label = "Cumulativa da distribuição de Laplace")

plt.legend(handles=[jardas_fra_plot, cdf_normal_plot, cdf_laplace_plot])

for i in range(len(jardas)):
    plt.plot((jardas[i], jardas[i]), (freq_acumulada[i], cdf_normal[i]), ':', color="red", alpha=0.3)
    #Omitido para não poluir muito o gráfico
    plt.plot((jardas[i], jardas[i]), (freq_acumulada[i], cdf_laplace[i]), '--', color="green", alpha=0.3)
plt.legend(loc='best')
```

De acordo com o gráfico do plot das cumulativas, o valor obtido pela distribuição da Normal é muito mais preciso do que os valores obtidos pela distribuição de Laplace. Isto ocorre, pois a distância dos dados para os valores nas distribuições é muito menor para a distribuição Normal.

### Itens 10 e 11 - Resolução:

```python
# Diferenças absolutas
soma_difs_cdf_normal = sum( abs(freq_acumulada - cdf_normal))
soma_difs_cdf_laplace = sum( abs(freq_acumulada - cdf_laplace))

print("Erro da soma do módulo das diferenças da Normal versus F.R.A: {0} \nErro da soma do módulo das diferenças da Laplace versus F.R.A: {1}".format(soma_difs_cdf_normal, soma_difs_cdf_laplace))
```

Como pode-se observar, o Erro da soma do módulo das diferenças entre a distribuição e a frequência acumulada de Jardas é menor na distribuição normal, portanto, a Normal é quem descreve melhor o modelo teórico.

```python
diferenca_normal = jardas - q_normal
diferenca_laplace = jardas - q_laplace

df = pd.DataFrame({'Jardas' : jardas,
                   'Normal' : q_normal,
                   'Diferença Normal (Jardas - Normal)': diferenca_normal,
                   'LaPlace' : q_laplace,
                   "Diferença LaPlace (Jardas - Laplace)": diferenca_laplace})
df
```
